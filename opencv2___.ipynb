```
import cv2
import matplotlib.pyplot as plt
```

```
# checking version
print(cv2.__version__)
```

```
# image reading
img=cv2.imread('mob.jpg',1)  # 1 means image in same color channel  --GRB or RGB
img1=cv2.imread('mob.jpg',0)# 0 means no color channel  -- black & white
```

```
print(img)
```

```
from google.colab.patches import cv2_imshow
```

```
cv2_imshow(img)
```

```
cv2_imshow(img1)
```

```
x,y,z=cv2.split(img)
```

```
cv2_imshow(x)
cv2_imshow(y)
cv2_imshow(z)
```

```
p1=(0,0)
p2=(150,160)
img2=cv2.line(img,p1,p2,(0,0,255),3)
```

```
cv2_imshow(img2)
```

```
encode_param=[int(cv2.cv2.IMWRITE_JPEG_QUALITY),90]
result, enimg=cv2.imencode('.jpg',img,encode_param)
```
